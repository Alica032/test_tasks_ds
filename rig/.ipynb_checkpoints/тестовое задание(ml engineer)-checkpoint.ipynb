## Задача
1. Вывести количество недостающих значений в поле 'duration'
2. Заменить недостающие значения в поле 'duration' медианой
3. В столбце 'genres' перечислены жанры, к которым относится фильм. Необходимо посчитать среднее количество жанров у фильмов
4. Найти фильм/фильмы с максимальным количеством жанров.
5. Найти самый часто встречаемый жанр.
6. Построить сводную таблицу(pivot). В ней должна содержаться информация о том, какой суммарный бюджет фильмов в разрезе стран за год, в 2014-2016 годах включительно.

```python
import pandas as pd
from functools import reduce
```

```python
df = pd.read_csv("movie.csv")
```

```python
df.head()
```

```python
df["duration"].isna().sum() #количество недостающих значений в поле 'duration'
```

```python
df.loc[df["duration"].isna(), "duration"] = df["duration"].median() #Замена недостающие значения в поле 'duration' медианой
```

```python
df["genres"]
```

```python
tags = map(lambda x: set(x.split("|")), df["genres"].unique())
tags = list(reduce(lambda x, y: x|y, tags, set()))
for tag in tags:
    df[tag] = df["genres"].apply(lambda x: 1 if tag in x else 0)
```

```python
df["mean_genres"] = df[tags].mean(axis=1) #среднее количество жанров у фильмов
```

```python
df[df["mean_genres"]==df["mean_genres"].max()] #фильм с максимальным количеством жанров - "Anastasia" 
```

```python
df[tags].sum().sort_values(ascending=False) # Drama - популярный жанр
```

```python
df["budget_count"] = df["budget"].apply(lambda x: float(x.split("$")[0]))
df_pivot = df.loc[(2016>=df["title_year"]) & (df["title_year"] >=2014)]
df_pivot = df_pivot.dropna() # под вопросом нули в бюджете, но на результат не повлияет
df_pivot.pivot_table(index='country', columns='title_year', values='budget_count', aggfunc='sum')
```
